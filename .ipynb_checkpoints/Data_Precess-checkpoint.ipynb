```python
# 加载数据分析常用库
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob, os
import seaborn as sns
import sys
from sklearn.preprocessing import MinMaxScaler

#--------------数据预处理----------------#

data = pd.read_csv('./data/dataCleaned.csv')

#1.处理时间特征，将时间特征处理为"星期“和”效率时间段“
df = pd.DataFrame(data)

#将医生编号改成人名，可以在创造数据时改，但那个太麻烦了，所以放在这里改
doctor = []
doc = df['Doctor']
for item in doc:
    if item == 1:
        doctor.append('Tom')
    elif item == 2:
        doctor.append("Mary")
df["Doctor"] = doctor

#将疾病编号改成疾病，可以在创造数据时改，但那个太麻烦了，所以放在这里改
disease_data = pd.read_csv("./data/data_clean.csv")
df2 = pd.DataFrame(disease_data)
df2 = df2['disease']
df2.drop_duplicates(keep='first',inplace=True)
df2 = df2.reset_index()
df2 = df2['disease']
#print(df2)
disease = []
dis = df['Disease']
for item in dis:
    disease.append(df2[item])

df["Disease"] = disease    


time = df['Time']
date = []
for item in time:
    date.append(pd.Timestamp(item))
    
#星期特征
week = []
for time in date:
    
    if time.dayofweek == 1:
        name = "Mondy"
    elif time.dayofweek == 2:
        name = "Tuesday"
    elif time.dayofweek == 3:
        name = "Wednesday"
    elif time.dayofweek == 4:
        name = "Thursday"
    elif time.dayofweek == 5:
        name = "Firday"
    week.append(name)


#效率时间段特征
type = []
for time in date:
    if (time.hour >= 8 and time.hour < 11) or (time.hour >= 14 and time.hour < 16):
        type.append("H")
    else:
        type.append("L")
df['Type'] = type
df['Week'] = week
df = df[['Doctor','Disease','Type','Week','WaitTime']]
df = pd.get_dummies(df)
df.to_csv("./data/data.csv")


#------------------获取训练数据--------------------#

```
